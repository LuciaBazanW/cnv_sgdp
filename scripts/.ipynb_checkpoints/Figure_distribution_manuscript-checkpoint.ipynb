```python
#pd.set_option('display.max_rows', 400)

data = pd.read_csv("../call_cnvr_grch37/individual_cnv_cnvr.txt", sep='\t')
anotation = pd.read_csv('../data/SGDP_anotation.csv', sep=',', encoding='latin-1',  index_col=1)

distribution = data[data['Length'] < 1000000]
distribution

### Plotiing distribution
duplications = distribution[distribution['CNV_Value']>1]
duplications['TYPE'] = 'duplication'
#duplications = duplications[(duplications[duplications.columns[3:-1]]<10).all(axis=1)]
deletions = distribution[distribution['CNV_Value']<3]
deletions['TYPE'] = 'deletion'


data_no_sex_chr= pd.concat([deletions,duplications])
data_no_sex_chr

data_no_sex_chr = data_no_sex_chr.pivot_table(index=["Chr", "Start", "End", "TYPE", 'Length'], 
                    columns='Sample_ID', 
                    values='CNV_Value').reset_index()
data_no_sex_chr
```

```python
chm13 = pd.read_csv('t2tchm13_cnvs.csv', index_col=0)
chm13['LENGHT'] = chm13['End'] - chm13['Start']
chm13 = chm13.drop_duplicates()


chm13 = chm13[chm13['LENGHT'] < 1000000] # 2000000
chm13 = chm13[chm13['LENGHT'] > 1000] # 
data
## Filter by chromosomes

data_no_sex_chr_chm13 = chm13[chm13.Chromosome != 'Y']
data_no_sex_chr_chm13 = data_no_sex_chr_chm13[data_no_sex_chr_chm13.Chromosome != 'X']
data_no_sex_chr_chm13 = data_no_sex_chr_chm13[data_no_sex_chr_chm13.Chromosome != 'MT']

### Plotiing distribution
duplications = chm13[(chm13[chm13.columns[3:-1]]>1).all(axis=1)]
duplications['TYPE'] = 'duplication'
#duplications = duplications[(duplications[duplications.columns[3:-1]]<10).all(axis=1)]
deletions = chm13[(chm13[chm13.columns[3:-1]]<3).all(axis=1)]
deletions['TYPE'] = 'deletion'


data_no_sex_chr_chm13= pd.concat([deletions,duplications])
data_no_sex_chr_chm13
```
