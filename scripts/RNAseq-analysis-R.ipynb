Data processing	Basecalls performed using Illumina CASAVA version 1.4
Reads were mapped to the human reference genome (hg19) using bowtie-2.0.0 and tophat-2.0.4 split read mapping algorithms using the “-b2-very-sensitive” parameters
Reads were filtered to include only properly paired reads.
Transcript level quantification was performed with cufflinks-2.0.2 to produce gene and transcript FPKM estimates
Genome_build: hg19
Supplementary_files_format_and_content: tab-delimited text files for gene and isoform FPKM values for each sample
