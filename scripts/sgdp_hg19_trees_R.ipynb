```r
library(tidyr)
library(dplyr)
library(readr)
library(basetheme)


deletions = read.csv("/branchinecta/jbazanwilliamson/cnv_analysis/cnv_sgdp/data/deletions_overlapped_archaic.csv")
duplications = read.csv("/branchinecta/jbazanwilliamson/cnv_analysis/cnv_sgdp/data/duplications_overlapped_archaic.csv")
anotation = read.csv("/branchinecta/jbazanwilliamson/SGDP_anotation.csv")

anotation = anotation[order(anotation$X3.Illumina_ID),]
anotation
```

```r
#deletions %>% replace(deletions == 'NA', 2)
#replace(deletions, deletions == 'NA', 0)
deletions[is.na(deletions)] <- 2
duplications[is.na(duplications)] <- 2
```

```r
drop <- c("X","Chr","Start", "End")
df = deletions[,!(names(deletions) %in% drop)]


df = t(df)
df
```

```r
library(ape)
```

```r
D <- dist.gene(df)
```

```r
length(D) 
```

```r
treeUPGMA  <- upgma(D)
treeNJ  <- NJ(D)
```

```r
tre <- hclust(D)
class(tre)
```

```r
tre <- ladderize(tre)
tre 
```

```r
plot(tre, cex = 0.6)
#title("A Simple NJ Tree")
```

```r
h_cluster <- hclust(D, method="ward.D",members = NULL) # method = average is used for UPGMA, members can be equal to NULL or a vector with a length of size D
plot(h_cluster, cex = 0.6)
```
