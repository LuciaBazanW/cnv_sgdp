```r
weightedmean <- function(scores, ranges, qranges)
{
    isects <- pintersect(ranges, qranges)
    sum(scores * width(isects)) / sum(width(isects))
}
```

```r
qreduceAssay(ra, query, simplifyReduce = weightedmean)
```
