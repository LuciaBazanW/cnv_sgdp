```python
## GRCh37
import pandas as pd
```

```python
hg19 = pd.read_csv('cnvator_data_sudmant_overlapped.csv')
hg19

## ADDING DUPLICATION/DELETION COLUMN 

deletions = hg19[hg19['RD'] < 2 ] 
deletions['type'] = 'DEL'
duplications= hg19[hg19['RD'] > 2] 
duplications['type'] = 'DUP'

frames = [deletions,duplications]

hg19 = pd.concat(frames)
hg19
```

```python
input_hg19_annotsv = pd.DataFrame()
input_hg19_annotsv['chr'] = hg19['Chr']
input_hg19_annotsv['start'] = hg19['Start']
input_hg19_annotsv['end'] = hg19['End']
input_hg19_annotsv['type'] = hg19['type']
input_hg19_annotsv['sample'] = hg19['Sample_ID']
input_hg19_annotsv
```

```python
input_hg19_annotsv.to_csv('hg19_input_annotsv.bed', sep='\t', header=False, index=False)
```
